```python
import pandas as pd
```

```python
penguins_raw = pd.read_csv('penguins_cleaned.csv')
```

```python
penguins_raw
```

```python
penguins = penguins_raw.drop(columns=['species'])
```

```python
penguins
```

```python
encode = ['sex','island']
```

## OPERACIONES  EN PANDAS
***
* Si es axis = 0 se añade o concatena a la fila.
* Si es axis = 1 se añade a la columna.

```python
for col in encode:
    dummy = pd.get_dummies(penguins[col],prefix=col)
    df    = pd.concat([penguins,dummy],axis=1)# axis = 0 se añade o concatena a la fila
    del   df[col]                                    #pero si es axis= 1 se añade a la columan
    
    
```
